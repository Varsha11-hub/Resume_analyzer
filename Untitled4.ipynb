```python
import pandas as pd
import spacy

mlp=spacy.load("en_core_web_sm")

#import libraries into the python program
#pd ->alias for pandas
#you use pandas to :
#load csv
#analyze data
#manipulate tables
#loads the english NLP model into memory
#nlp becomes the NLP processing engine
#example usage:
#text ->nlp(text) -> AI analysis
```

```python
import pandas as pd
df=pd.read_csv("/content/Resume.csv",engine="python")
df.head()

#engine = "python" helps read large text fields like resumes.
```

```python
def extract_entites(text):

#defines a function that extracts entities from text
#input resume text
#output names
#organization

  doc=mlp(text)

#runs the resume text through the spac model
#AI analyze
#grammar
#entites
#names
#organisation

  names=[]
  orgs=[]

#creates empty lines to stores:
#name and orgs

  for ent in doc.ents:
    if ent.label_=="PERSON":
      names.append(ent.text)
#if the entity is a person name:
#add it to the names list
#example:
#John Smith ->person
#output :
#names=["john smith"]
    if ent.label_=="ORG":
      orgs.append(ent.text)
#if the entity is an organisation:
#add it to the organizaton list:
#example:
#Google ->ORG
#Output:
#orgs=['Google']
  return names ,orgs
```

```python
for text in df['Resume_str'].head(5):
  names,orgs=extract_entites(text)

  print("Names:", names)
  print("Organization:",orgs)
  print("-------------------")
```
